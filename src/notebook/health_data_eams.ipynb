```python
#flll NaN in additional diagnosis with the principal diagnosis
# def fill_na_with_principal(df):
#     df.fillna({'Additional Diagnosis (New Case)': df['Principal Diagnosis (New Case)']}, inplace=True)
#     #convert age to integers
#     df['Age'] = df['Age'].str.replace(' Year(s)', '').astype(int)
#     return df

def fill_na_with_principal(df):
    df.fillna({'Additional Diagnosis (New Case)': df['Principal Diagnosis (New Case)']}, inplace=True)
    
    # Convert age to integers - handle both years and months
    def convert_age_to_years(age_str):
        if pd.isna(age_str):
            return 0
        age_str = str(age_str).strip()
        if 'Month(s)' in age_str:
            months = int(age_str.replace(' Month(s)', ''))
            return round(months / 12, 1)  # Convert months to years
        elif 'Year(s)' in age_str:
            return int(age_str.replace(' Year(s)', ''))
        else:
            try:
                return int(age_str)
            except:
                return 0
    
    df['Age'] = df['Age'].apply(convert_age_to_years)
    return df

# convert binary data to integers
def convert_binary_to_integers(df):
    df.fillna({'Sex': df['Sex'].mode()[0]}, inplace=True)
    df['Sex'] = df['Sex'].str.lower().map({'male': 0, 'female': 1}).astype(int)
    df['Pregnant Patient'] = df['Pregnant Patient'].map({'Yes': 1, 'No': 0}).astype(int)
    df['NHIA Patient'] = df['NHIA Patient'].map({'Yes': 1, 'No': 0}).astype(int)
    return df
```

```python
new_malaria_df = fill_na_with_principal(new_malaria_df)
new_malaria_df = convert_binary_to_integers(new_malaria_df)
new_malaria_df.head()

new_diabetes_df = fill_na_with_principal(new_diabetes_df)
new_diabetes_df = convert_binary_to_integers(new_diabetes_df)
new_diabetes_df.head()
```

Get google trends data in relation to the one i got from weija hospital

```python
# --- Google Trends Fetch with Delay and Error Handling ---
import pandas as pd
from pytrends.request import TrendReq
import time
import os

months = [
    '2022-03', '2022-06', '2022-09', '2022-10',
    '2023-03', '2023-06', '2023-09', '2023-10',
    '2024-03', '2024-06', '2024-09', '2024-10'
]

pytrends = TrendReq(hl='en-US', tz=0)
kw_list = ['diabetes', 'malaria']

results = []
progress_file = '../artifacts/time_series/google_trends_time_series.csv'

# Load progress if exists
if os.path.exists(progress_file):
    results = pd.read_csv(progress_file).to_dict('records')
    done_months = set(r['Month'] for r in results)
else:
    done_months = set()

for month in months:
    if month in done_months:
        print(f"Skipping {month}, already fetched.")
        continue
    start = f"{month}-01"
    end = pd.to_datetime(start) + pd.offsets.MonthEnd(0)
    timeframe = f"{start} {end.strftime('%Y-%m-%d')}"
    try:
        pytrends.build_payload(kw_list, timeframe=timeframe, geo='')
        df = pytrends.interest_over_time()
        if not df.empty:
            diabetes_trend = df['diabetes'].mean()
            malaria_trend = df['malaria'].mean()
        else:
            diabetes_trend = malaria_trend = None
        results.append({'Month': month, 'diabetes_trend': diabetes_trend, 'malaria_trend': malaria_trend})
        # Save progress after each fetch
        pd.DataFrame(results).to_csv(progress_file, index=False)
        print(f"Fetched and saved {month}")
    except Exception as e:
        print(f"Error for {month}: {e}")
    time.sleep(60)  # Wait 60 seconds between requests

trends_df = pd.DataFrame(results)
print(trends_df)
```
