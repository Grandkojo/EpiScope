Exploring The Dataset Collected, for our Epidemic Analysing and Monitoring System for Malaria and Diabetes

```python
import pandas as pd
```

```python
df = pd.read_csv('../artifacts/csv/health_data_eams_diabetes.csv')
df.info()
```

```python
# remove columns that have little or no information
new_df = df.drop(columns=["Sr.No.", "Provisional Diagnosis", "Type Of Test(s) Requested", "Test Result(s)", "Principal Diagnosis (Old Case)", "Principal Diagnosis (Recurring Case)", "Additional Diagnosis (Old Case)", "Additional Diagnosis (Recurring Case)"])
new_df.head()
# new_df.info()
# new_df.isnull().sum()
# new_df.describe()
# new_df.describe(include=['O'])
# new_df.describe(include=['object'])
```

```python
#flll NaN in additional diagnosis with the principal diagnosis
new_df.fillna({'Additional Diagnosis (New Case)': new_df['Principal Diagnosis (New Case)']}, inplace=True)
#convert age to integers
new_df['Age'] = new_df['Age'].str.replace(' Year(s)', '').astype(int)
```

```python
# convert binary data to integers
new_df.fillna({'Sex': new_df['Sex'].mode()[0]}, inplace=True)
new_df['Sex'] = new_df['Sex'].str.lower().map({'male': 0, 'female': 1}).astype(int)
new_df['Pregnant Patient'] = new_df['Pregnant Patient'].map({'Yes': 1, 'No': 0}).astype(int)
new_df['NHIA Patient'] = new_df['NHIA Patient'].map({'Yes': 1, 'No': 0}).astype(int)

new_df.head()
```
